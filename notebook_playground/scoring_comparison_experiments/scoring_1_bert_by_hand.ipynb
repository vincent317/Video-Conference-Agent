```python
from evaluate import load
import os
bertscore = load("bertscore")
import warnings
warnings.filterwarnings("ignore")
```

```python
def bertscore_calc(predictions, references):
    
    num_diff = len(references) - len(predictions)
    assert(num_diff == 0)
    
#     # if we have more predictions than references, ignore the right predictions
#     if num_diff < 0:
#         predictions = predictions[:len(references)]
        
#     # if we have less predictions than references, bad
#     if num_diff > 0:
#         for _ in range(num_diff):
#             predictions.append("")
     
    results = bertscore.compute(predictions=predictions, references=references, lang="en")
        
    
    return (results['f1'], sum(results['f1']) / len(results['f1']))
```
